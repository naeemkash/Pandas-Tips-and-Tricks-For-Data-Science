# 16- Splitting a string into multiple column

```
# import libraries
import pandas as pd
df = pd.DataFrame({"Name":['Muhammad Naeem', "Muhammad Waseem",'Muhammad Nadeem','Zeshan Ali'],
                   'Location':['karachi Pakistan' , 'islamabad Pakistan' , 'Lohor Pakistan', 'Hydrabad Pakistan']})
df
```

```
# spliting a column into two columns
df.Name.str.split(' ', expand=True)
```

```
# addinng those split into new columns
df[["first_name", "Last_name"]]= df.Name.str.split(' ', expand=True)
df
```

```
df[["city", "country"]]= df.Location.str.split(' ', expand=True)
df
```

```
# Refine Data manipulation
df = df[['first_name' , 'Last_name','city','country']]
df
```
